### 2.2. Estrategia Climática Nacional de México y Contribuciones Determinadas a Nivel Nacional (NDCs)
En consonancia con los esfuerzos globales, México ha asumido compromisos significativos para enfrentar el cambio climático, plasmados principalmente en sus Contribuciones Determinadas a Nivel Nacional (NDCs, por sus siglas en inglés). En su actualización más reciente de 2022, México se comprometió a una meta de reducción no condicionada de sus emisiones de gases de efecto invernadero (GEI) del 35% para el año 2030, con respecto a un escenario base o tendencial. Adicionalmente, estableció una meta condicionada de reducción del 40% para el mismo año, sujeta a la disponibilidad de cooperación y financiamiento internacional. Estos compromisos son de amplio alcance y abarcan sectores clave de la economía, incluyendo energía, procesos industriales, transporte, agricultura, uso de suelo, cambio de uso de suelo y silvicultura (USCUSS) y residuos. A más largo plazo, la estrategia nacional de México contempla alcanzar la neutralidad de carbono, es decir, un equilibrio entre las emisiones emitidas y las absorbidas, para mediados de siglo. La NDC actualizada representa un incremento en la ambición en comparación con la meta previa del 22% de reducción, señalando una mayor determinación del país para contribuir a los objetivos del Acuerdo de París.   

Para alcanzar estas metas, México ha delineado una serie de políticas y medidas específicas. En el sector energético, se planea una mayor integración de fuentes de energía limpia, incluyendo la modernización y repotenciación de centrales hidroeléctricas, el aumento de la capacidad de generación fotovoltaica, eólica y geotérmica, y el fomento de la generación distribuida renovable. También se impulsarán nuevas tecnologías como el hidrógeno verde en centrales híbridas. Otras medidas incluyen la sustitución de combustibles de alto contenido de carbono por gas natural en centrales de alta eficiencia y la reducción de las pérdidas técnicas en la red eléctrica nacional. Específicamente para el sector de petróleo y gas, se ha establecido una meta de reducción de emisiones del 14%, que se buscará alcanzar mediante el incremento de la cogeneración, la reducción de emisiones fugitivas y otros programas de eficiencia. El escenario tendencial de emisiones proyectado para 2030, sin la implementación de estas políticas de mitigación, se cuantificó en 991 millones de toneladas de CO2 equivalente (MtCO2e), cifra que sirve como línea base para medir el esfuerzo de reducción.   

La existencia de una meta de reducción condicionada, que representa un 5% adicional de disminución de emisiones dependiente de la cooperación y el financiamiento internacional , abre una ventana de oportunidad para la colaboración y el apoyo externo. Un análisis prospectivo de las emisiones, como el que se propone en este estudio, puede ser valioso para cuantificar la brecha entre la trayectoria esperada bajo las políticas actuales y los recursos nacionales, y la trayectoria necesaria para alcanzar la meta condicionada. ## 2. Proceso de Extracción y Limpieza de Datos
La preparación adecuada de los datos es un paso fundamental para cualquier análisis de series de tiempo. Este proceso asegura que los datos estén en un formato manejable y libre de inconsistencias que puedan afectar la validez de los modelos y los pronósticos subsecuentes.

### 2.1. Carga de Datos y Selección de la Serie para México
El primer paso consistió en cargar el conjunto de datos CO2 Emissions.xls - Data.csv  en un entorno de análisis de datos utilizando la biblioteca pandas de Python. Este archivo contiene las emisiones de CO_2 per cápita para múltiples países y regiones a lo largo de varios años.   

Una vez cargados los datos, se procedió a filtrar la información para aislar específicamente la serie temporal correspondiente a México. Esto se realizó identificando a México por su nombre ("Mexico") y su código de país ("MEX"). Se seleccionó el indicador "CO2 emissions (metric tons per capita)" (código de indicador: EN.ATM.CO2E.PC).

El formato original de los datos es "ancho", donde cada año representa una columna. Para el análisis de series de tiempo, es más conveniente un formato "largo", donde una columna representa el tiempo (año) y otra el valor de la emisión. Esta transformación es crucial para la compatibilidad con las bibliotecas de modelado de series temporales en Python, como statsmodels.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Iniciamos con análisis exploratorio de los datos, para descubrir si es necesario aplicar alguna transformación o transformación a los datos
df_raw = pd.read_csv('CO2_emissions_mexico.csv')

# Filtrar por México y el indicador específico
df_mex_raw = df_raw[(df_raw['CountryName'] == 'Mexico') & (df_raw['IndicatorCode'] == 'EN.GHG.CO2.PC.CE.AR5')]

# Seleccionar columnas relevantes (años) y transformar a formato largo
years_columns = [str(year) for year in range(1970, 2024)]
df_mex_long = df_mex_raw.melt(id_vars=['CountryName', 'IndicatorName'],
                                value_vars=years_columns,
                                var_name='Year',
                                value_name='CO2_per_capita')

print(df_mex_long.head())
print(df_mex_long.tail())
```

```python

try:
    df_total = pd.read_csv('CO2_emissions_mexico.csv')
except FileNotFoundError:
    print("Error: El archivo 'CO2_emissions_mexico.csv' no se encontró. Por favor, verifica la ruta.")
    exit()

# Filtrar datos para México
df_mexico_raw = df_total[df_total['CountryCode'] == 'MEX'].copy()

# Seleccionar las columnas de años relevantes (1970-2019) y el CountryCode
cols_years = [str(year) for year in range(1970, 2020)]
df_mexico_emissions = df_mexico_raw[['CountryCode'] + cols_years]

# Transponer el DataFrame para tener años como filas
df_mexico_emissions_t = df_mexico_emissions.set_index('CountryCode').transpose()

# Renombrar la columna a 'Emissions_CO2_per_capita'
df_mexico_emissions_t.rename(columns={'MEX': 'Emissions_CO2_per_capita'}, inplace=True)

# Convertir la columna de emisiones a tipo numérico
df_mexico_emissions_t['Emissions_CO2_per_capita'] = pd.to_numeric(df_mexico_emissions_t['Emissions_CO2_per_capita'], errors='coerce')

# Convertir el índice a tipo datetime (Año)
# Se crea un índice de Periodo para asegurar que la frecuencia sea anual y evitar problemas con días/meses
df_mexico_emissions_t.index = pd.to_datetime(df_mexico_emissions_t.index, format='%Y').to_period('A')
df_mexico_emissions_t.index.name = 'Year'


# Eliminar filas con valores NaN en la columna de emisiones (si las hubiera después de la conversión)
df_mexico_series = df_mexico_emissions_t[['Emissions_CO2_per_capita']].dropna()

# Calcular estadísticas descriptivas
descriptive_stats = df_mexico_series['Emissions_CO2_per_capita'].describe()
min_val = df_mexico_series['Emissions_CO2_per_capita'].min()
min_year = df_mexico_series['Emissions_CO2_per_capita'].idxmin().year
max_val = df_mexico_series['Emissions_CO2_per_capita'].max()
max_year = df_mexico_series['Emissions_CO2_per_capita'].idxmax().year
q1 = df_mexico_series['Emissions_CO2_per_capita'].quantile(0.25)
q3 = df_mexico_series['Emissions_CO2_per_capita'].quantile(0.75)
iqr = q3 - q1

print("Estadísticas Descriptivas para México (1970-2019):")
print(f"Media: {descriptive_stats['mean']:.2f} t CO2e/cápita")
print(f"Mediana: {descriptive_stats['50%']:.2f} t CO2e/cápita")
print(f"Desviación Estándar: {descriptive_stats['std']:.2f} t CO2e/cápita")
print(f"Mínimo: {min_val:.2f} t CO2e/cápita (en {min_year})")
print(f"Máximo: {max_val:.2f} t CO2e/cápita (en {max_year})")
print(f"Rango Intercuartílico: {iqr:.2f} t CO2e/cápita")
```

### 2.2. Preparación de la Serie Temporal
Una vez extraída la serie para México, se realizaron los siguientes pasos de preparación:

- Conversión del Año a Formato de Fecha: La columna 'Year', que inicialmente contenía cadenas de texto representando los años, se convirtió a objetos datetime de pandas. Para datos anuales, es común asignar el primer día del año para representar cada observación temporal (e.g., '1960' se convierte a 1960-01-01).
- Establecimiento del Índice Temporal: La columna 'Year' (ahora en formato datetime) se estableció como el índice del DataFrame. Esto es un requisito para la mayoría de las funciones de análisis de series de tiempo en statsmodels.
- Manejo de Valores Faltantes: Se inspeccionó la serie para detectar valores faltantes (NaN). El archivo  muestra datos para México hasta 2019, con valores para 2020 y 2021, y columnas vacías para 2022 y 2023 en la vista general. Para este análisis, se utilizarán los datos hasta el último año con información no nula disponible para México, que es 2021. Se verificó que no existieran valores faltantes intermitentes dentro de este rango.   
- Verificación del Tipo de Dato: Se aseguró que la columna 'CO2_per_capita' fuera de tipo numérico (flotante) para permitir cálculos y modelado.

```python
# Convertir 'Year' a datetime y 'CO2_per_capita' a numérico
df_mex_long['Year'] = pd.to_datetime(df_mex_long['Year'], format='%Y')
df_mex_long['CO2_per_capita'] = pd.to_numeric(df_mex_long['CO2_per_capita'], errors='coerce')

# Establecer 'Year' como índice
df_mex_ts = df_mex_long.set_index('Year')[['CO2_per_capita']]

# Eliminar filas con NaN en CO2_per_capita (esto manejará los años finales sin datos)
df_mex_ts.dropna(subset=['CO2_per_capita'], inplace=True)

# Verificar la serie temporal procesada
print("Primeras observaciones:")
print(df_mex_ts.head())
print("\nÚltimas observaciones:")
print(df_mex_ts.tail())
print("\nInformación del DataFrame:")
print(df_mex_ts.info())
print("\nEstadísticas descriptivas básicas:")
print(df_mex_ts.describe())
```

## 3.2. Descomposición de la Serie Temporal
Para obtener una visión más clara de los componentes subyacentes de la serie, se aplicó una descomposición clásica. Dado que los datos son anuales, la "estacionalidad" en el sentido tradicional (patrones mensuales o trimestrales) no es aplicable. Sin embargo, la descomposición puede ayudar a aislar la tendencia a largo plazo y los componentes cíclicos o irregulares. Se optó por un modelo de descomposición aditivo, asumiendo que la magnitud de cualquier componente cíclico o irregular no depende del nivel de la tendencia. La teoría y ejemplos de descomposición se describen en las notas del curso.   



```python
# Descomposición de la serie temporal (aditiva, ya que la estacionalidad anual no es típica)
# Para datos anuales, 'period' sería 1 si no se esperan ciclos específicos,
# o un valor mayor si se sospechan ciclos económicos de cierta duración (ej. 5-10 años).
# Aquí, se usará un enfoque más simple para visualizar la tendencia y los residuos.
decomposition = seasonal_decompose(df_mex_ts['CO2_per_capita'], model='additive', period=1) # period=1 para datos anuales sin estacionalidad

trend_component = decomposition.trend
seasonal_component = decomposition.seasonal # Para datos anuales con period=1, esto será cero.
residual_component = decomposition.resid

# Gráficos de los componentes descompuestos
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(df_mex_ts['CO2_per_capita'], label='Observado')
plt.legend(loc='upper left')
plt.title('Descomposición de la Serie Temporal de Emisiones de CO2 per cápita en México')

plt.subplot(412)
plt.plot(trend_component, label='Tendencia', color='orange')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(seasonal_component, label='Estacionalidad/Ciclo', color='green') # Será una línea en cero para period=1
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(residual_component, label='Residual', color='red')
plt.legend(loc='upper left')
plt.xlabel('Año')

plt.tight_layout()
plt.show()
```

## 4. Pruebas de Identificación del Modelo
Antes de implementar modelos de series de tiempo como ARIMA, es crucial identificar adecuadamente las características de la serie, especialmente su estacionariedad y la estructura de sus autocorrelaciones.

4.1. Análisis de Estacionariedad
Un proceso estacionario es aquel cuyas propiedades estadísticas (como la media, la varianza y la autocovarianza) no cambian con el tiempo. Específicamente, se trabaja bajo el supuesto de estacionariedad débil, que requiere:

Una media constante: μ_t =μ para todo t.
Una función de autocovarianza que dependa solo del rezago h: γ(s,t)=γ(∣t−s∣) para cualquier s,t. La mayoría de los modelos ARIMA asumen que la serie es estacionaria o se transforma para lograr la estacionariedad (generalmente a través de la diferenciación).   
Para evaluar formalmente la estacionariedad de la serie de emisiones de CO_2 per cápita de México, se utilizó la prueba Aumentada de Dickey-Fuller (ADF). Las hipótesis de esta prueba son :   
H_0: La serie tiene una raíz unitaria (es decir, es no estacionaria).
H_a: La serie no tiene una raíz unitaria (es decir, es estacionaria).
Se busca un valor p pequeño (típicamente <0.05) para rechazar la hipótesis nula y concluir que la serie es estacionaria.



```python


# Prueba Aumentada de Dickey-Fuller (ADF) en la serie original
adf_result_original = adfuller(df_mex_ts['CO2_per_capita'].dropna())

# Acceder a los resultados individuales
adf_statistic = adf_result_original[0]
p_value = adf_result_original[1]
critical_values = adf_result_original[4] # Este es el diccionario

print(f'ADF Statistic (Original): {adf_statistic}')
print(f'p-value (Original): {p_value}')
print('Critical Values (Original):')
for key, value in critical_values.items(): # Ahora sí, .items() sobre el diccionario
    print(f'\t{key}: {value}')

# Interpretación
print("\nInterpretación:")
if p_value > 0.05:
    print(f"El p-valor es {p_value:.4f}. La serie original es NO ESTACIONARIA (no se rechaza H0).")
else:
    print(f"El p-valor es {p_value:.4f}. La serie original es ESTACIONARIA (se rechaza H0).")
```
